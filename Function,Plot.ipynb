# y=ax^2+bx+c

```python
def quad(x):
    a=2
    b=-4
    c=10
    y=a*x**2+b*x+c
    return (y)
```

```python
X= [] #empty list
Y= []
for x in range(100):
    X.append(x)
    y1 = quad(x)
    Y.append(y1)
print(Y)
```

# Plot sine,cosine,tan,linear and quarditic plot

```python
import numpy as np #np is arbitary
X=[] #empty list to collect x values
S=[] #empty list to collect sinx values
C=[] #empty list to collect cosx values
T=[] #empty list to collect tanx values
L=[] #empty list to collect linear function values
Q=[] #empty list to collect quarditic function values
for x in range(100): 
    X.append(x) 
    ss = np.sin(x)
    cc = np.cos(x)
    tt = np.tan(x)
    ll = 5*x-6
    qq = 2*x**2-4*x+10
    S.append(ss)
    C.append(cc)
    T.append(tt)
    L.append(ll)
    Q.append(qq)    
```

```python
print (X,S,C,T,L,Q)
```

```python
plt.plot(X,S)
```

```python
plt.plot(X,C)
```

```python
plt.plot(X,T)
```

```python
plt.plot(X,Q) 
```

```python
plt.plot(X,L) 
```

```python
plt.plot(X,S)
plt.plot(X,C)
plt.plot(X,T)
plt.plot(X,L)
plt.plot(X,Q)
```
