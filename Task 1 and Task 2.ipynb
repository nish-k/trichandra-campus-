# Task-1

Gravitational force between two masses sepearated some distance $F=G\frac{m_1 m_2}{r^2}$

Gravitational potential of mass at some distance $V=G\frac{m_1}{r}$

Gravitational intensity of a mass at some diatance $g=G\frac{m_1}{r^2}$

# -Create class of these three functions.

```python
class Kg(): 
    def __init__ (self,G,m1,m2,r):
        self.G = G
        self.m1 = m1
        self.m2 = m2
        self.r = r
    def Force(self,r):
        F=self.G*self.m1*self.m2/self.r**2 
        return F 
    def Potential(self,r):
        V=self.G*self.m1/self.r
        return V
    def Intensity(self,r): 
        g=self.G*self.m1/self.r**2
        return g
        
```

```python
G=6.67*10e-11
m1 = 9.1*10e-31 #in Kg
m2 = 1.67*10e-27 # in Kg
M=[] #empty list to collect distance r
F=[] #empty list to collect force
V=[] #empty list to collect potential
g=[] #empty list to collect intensity
for r in range(10,1000): #for loop
    C1= Kg(G,m1,m2,r)
    ff=C1.Force(r)
    vv=C1.Potential(r)
    gg=C1.Intensity(r)
    M.append(r)
    F.append(ff)
    V.append(vv)
    g.append(gg)
   #print(r)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.figure(figsize = [10,10])
plt.subplot(2,2,1)
plt.plot(F,M)
plt.plot(V,M)
plt.plot(g,M) 
```

```python
plt.figure(figsize = [10,10])
plt.subplot(2,2,2)
plt.semilogy(M, F, label = 'force')
plt.semilogy(M, V, label = 'potential')
plt.semilogy(M, g, label = 'intensity')
plt.legend()
plt.xlabel(r)
plt.ylabel('ln F, ln V,ln g')

plt.subplot(2,2,3)
plt.loglog(M, F, label = 'force')
plt.loglog(M, V, label = 'potential')
plt.loglog(M, g, label = 'intensity')
plt.legend()
plt.xlabel('ln r')
plt.ylabel('ln F, ln V, ln g')
```

4.Plot the variation of electrostatic and gravitational force between them using matplotlib..

```python
class Variation():
    def __init__(self,G,K,m1,m2,q1,q2):
        self.G = G
        self.K = K
        self.m1 = m1 
        self.m2 = m2
        self.q1 = q1
        self.q2 = q2
    def f1(self,r):
         F1 = (self.G*self.m1*self.m2)/(r**2)
         return F1
    def f2(self,r):
        F2 = (self.K*self.q1*self.q2)/(r**2)
        return(F2)
```

```python
V=Variation(6.67e-11,9e9,9.1e-31,1.67e-27,1.6e-19,1.6e-19)
R=[]
M=[]
N=[]
for r in range (1,1000):
    g=V.f1(r)
    e=V.f2(r)
    R.append(r) 
    M.append(g)
    N.append(e)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = [10,10])
plt.subplot(2,2,1)
plt.plot(R,M, label = "Electrostatic Force")
plt.plot(R,N, label = "Gravitational Force")
```
